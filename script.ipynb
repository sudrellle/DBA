# [dbo].[fn_stock_restant]

```sql
/****** Object:  UserDefinedFunction [dbo].[fn_stock_restant]    Script Date: 02/06/2025 08:59:32 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE FUNCTION [dbo].[fn_stock_restant] (@id_produit INT)
RETURNS INT
AS
BEGIN
  DECLARE @stock_restant INT
  SELECT @stock_restant = 
    (SELECT stock_initial - ISNULL(SUM(quantite_achat),0)
     FROM Achat
     WHERE id_produit = @id_produit)
  FROM Produit
  WHERE id = @id_produit
  RETURN @stock_restant
END;
GO
```

# [dbo].[V_detail_client]

```sql
/****** Object:  View [dbo].[V_detail_client]    Script Date: 02/06/2025 08:59:32 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[V_detail_client] AS
WITH achats_clients AS (
    SELECT 
        c.nom_client,
        COUNT(a.id_client) AS nombre_achat,
        SUM(a.quantite_achat) AS total_quantite,
        SUM(p.prix * a.quantite_achat) AS total_depense
    FROM achat a
    JOIN client c ON a.id_client = c.id
    JOIN produit p ON a.id_produit = p.id
    GROUP BY c.nom_client
	
)
SELECT 
    nom_client,
    nombre_achat,
    total_quantite,
    total_depense,
    RANK() OVER (ORDER BY total_quantite DESC) AS rang_somme_quantite,
    RANK() OVER (ORDER BY nombre_achat DESC) AS rang_fidelisation,
	RANK () OVER(ORDER BY total_depense DESC) as rang_revenu
FROM achats_clients;
GO
```

# [dbo].[nb_tranche_age]

```sql
/****** Object:  View [dbo].[nb_tranche_age]    Script Date: 02/06/2025 08:59:32 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[nb_tranche_age] AS
WITH tranche_cte AS (
    SELECT 
        CASE 
            WHEN c.age BETWEEN 15 AND 25 THEN '15-25ans'
            WHEN c.age BETWEEN 26 AND 40 THEN '26-40ans'
            WHEN c.age BETWEEN 41 AND 60 THEN '41-60ans'
            ELSE '60+ans'
        END AS tranche_age,
        a.quantite_achat,
        p.prix
    FROM produit p
    JOIN achat a ON p.id = a.id_produit
    JOIN client c ON a.id_client = c.Id
)
SELECT 
    tranche_age,
    SUM(quantite_achat) AS SommeQuantite,
    SUM(prix * quantite_achat) AS revenugenere
FROM tranche_cte
GROUP BY tranche_age;











GO
```

# [dbo].[topProduitMensuel]

```sql
/****** Object:  View [dbo].[topProduitMensuel]    Script Date: 02/06/2025 08:59:32 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[topProduitMensuel] AS
WITH sous_vue AS (
    SELECT 
        DATENAME(MONTH, a.date_achat) AS mois,
        p.nom_produit,
        SUM(a.quantite_achat) AS total_quantite,
        ROW_NUMBER() OVER (
            PARTITION BY DATENAME(MONTH, a.date_achat)
            ORDER BY SUM(a.quantite_achat) DESC
        ) AS rang
    FROM produit p
    JOIN achat a ON p.id = a.id_produit
    GROUP BY 
        DATENAME(MONTH, a.date_achat),
        p.nom_produit
)
SELECT 
    mois,
    nom_produit,
    total_quantite
FROM sous_vue
WHERE rang = 1
GO
```
